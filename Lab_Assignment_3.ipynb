```python
class Node:
    def __init__(self,element,next,prev):
        self.element=element
        self.next=next
        self.prev=prev
class DoublyList:
    def __init__(self,a):
        self.head=Node(None,None,None)
        self.head.next=self.head.prev=self.head
        tail=None
        for i in a:
            newNode=Node(i,None,None)
            if self.head.next==self.head:
                self.head.next=newNode
                newNode.prev=self.head
                newNode.next=self.head
                self.head.prev=newNode
                tail=newNode
            else:
                tail.next=newNode
                newNode.prev=tail
                newNode.next=self.head
                self.head.prev=newNode
                tail=newNode
#task 2
    def showList(self):
        n=self.head.next
        while n!=self.head:
            print(n.element)
            n=n.next
#task(3,4)
    def insert(self,newElement,index=None):
        node=Node(newElement,None,None)
        size = 0
        n = self.head.next
        while n != self.head:
            size += 1
            n = n.next
        if index==None:
            n=self.head.next
            while n.next!=self.head:
                n=n.next
            temp=n.next
            n.next=node
            node.prev=n
            node.next=temp
        else:
            if index<0 or index>size:
                raise Exception("Invalid index")
            if index==0:
                temp=self.head.next
                self.head.next=node
                node.next=temp
                temp.prev=node
            else:
                if index==size:
                    a=self.head.next
                    while a.next!=self.head:
                        a=a.next
                    temp=a.next
                    a.next=node
                    node.prev=a
                    node.next=temp
                    temp.prev=a
                else:
                    a = self.head.next
                    i = 0
                    while i <= size:
                        if i == index - 1:
                            prev = a
                        i += 1
                    temp = prev.next
                    prev.next = node
                    node.next = temp
                    node.prev = prev
                    temp.prev = node
#task5
    def remove(self,index):
        size = 0
        n = self.head.next
        while n != self.head:
            size += 1
            n = n.next
        if index<0 or index>size:
            raise Exception("Invalid Index")
        if index==0:
            temp=self.head.next
            self.head.next=temp.next
            temp.next.prev=self.head
            temp.element=None
            temp.next=None
            temp.prev=None
        else:
            if index==size-1:
                a=self.head.next
                c=0
                while c<=size-2:
                    a=a.next
                    c+=1
                a.next = self.head
                self.head.prev = a
                a.element = None
                a.next = None
                a.prev = None
            else:
                c=0
                a=self.head.next
                while a!=self.head:
                    if c==index-1:
                        prev=a
                    if c==index:
                        removeNode=a
                    a=a.next
                    c+=1
                temp=removeNode.next
                prev.next=temp
                temp.prev=prev
                removeNode.element=removeNode.prev=removeNode.next=None
#task6
    def removekey(self,elem):
        size = 0
        n = self.head.next
        while n != self.head:
            size += 1
            n = n.next
        a = self.head.next
        while a != self.head:
            if a.element==elem:
                back=a.prev
                remove=a
                back.next = remove.next
                remove.next.prev = back
                return a.element
            a = a.next
a=[1,2,3,4]
h=DoublyList(a)
h.insert(5,4)
h.showList()
h.remove(3)
h.showList()
print(h.removekey(2))
h.showList()
```
